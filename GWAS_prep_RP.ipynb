```python
#####This step has many readlines, it outputs all variants in the cold spot

def filter_vcf_prep(): 
    fileIN = open(paramDict["gtexDir"] +'/chr22_subset_gtex_copy2.vcf')
    outlist = list()

    
    for i in fileIN:
        if i[0]=="#": continue
        i = i.rstrip().split('\t')
        if i[1]=="POS":
            continue      
        chrnum = str(i[0])
        #print(i[1])
        position = float(i[1])
        snp = str(i[2])
        outlist.append([chrnum, position, snp])

    return outlist  
    fileIN.close()

filter_vcf_prep()
```

```python
def add_pos_gwas(outfilename):
    fileIN = open('/users/michaelbinkley/desktop/GTEx/GWAS_RP.txt')
    fileOUT = open(outfilename, "w")
    vcf = filter_vcf_prep()
    outlist2 = list()
    for i in fileIN:
        
        i = i.rstrip().split('\t')
        
        for  (chrnum), (position), (snp) in vcf:
            if chrnum  != i[2]:
                continue
            if  position == float(i[3]):
                    outlist2.append( [str(snp), i[1], i[2], i[3], i[4], i[5], i[6], i[7], i[8], i[9]])


    fileOUT.write("\t".join([str(x) for x in outlist2] ) + "\n")
    fileIN.close()
    fileOUT.close()
add_pos_gwas('gwas_RP_complete.txt')
```

```python
def hapmap_vector():
    fileIN = open('/users/michaelbinkley/desktop/hapmap3_r1_b36_fwd.CEU.qc.poly.recode.map')
    hapmap = list()
    for i in fileIN: 
        i = i.rstrip().split('\t')
        Chr = str(i[0])
        rs = str(i[1])
        pos = str(i[3])
        hapmap.append([rs, Chr, pos])
    return hapmap
    fileIN.close()
hapmap_vector()
```

```python
def process_ldsc_input(outfilename):
    fileIN = open('/users/michaelbinkley/desktop/tsbca2.csv')
    fileIN.readline()
    fileOUT = open(outfilename, "w")
    hapmap = hapmap_vector()
    for rs, Chr, pos in hapmap:
        pos2 = float(pos)
    for i in fileIN:
        i = i.rstrip().split(',')   
        RS = str(i[0])
        CHR = str(i[1])
        POS = float(i[2])
        a1 = i[3]
        a2 = i[4]
        OR = i[5]
        se = i[6]
        pvalue = i[7]
        for rs, Chr, pos in hapmap:
            if RS == rs: 
                fileOUT.write(str(RS) + "\t" + '0' + '\t'+ str(RS) + "\t"+ str(CHR) + "\t" + str(POS) + "\t" + str(a1)+ "\t" + str(a2) + "\t" + str(OR) + "\t" + str(se) + "\t" + str(pvalue)  + "\n")
            else:  
                if CHR == Chr and abs((POS) - float(pos)) <160000: 
                    fileOUT.write(str(RS) + "\t" + str(abs((POS) - float(pos))) + '\t'+ str(rs) + "\t" + str(Chr) + "\t" + str(pos) + "\t" + str(a1)+ "\t" + str(a2) + "\t" + str(OR) + "\t" + str(se) + "\t" + str(pvalue) +  "\n")
    fileIN.close()
    fileOUT.close()
process_ldsc_input("HapMap_SNPlist.txt")    
```

```python
def output_ldsc():
    fileIN = open("HapMap_SNPlist.txt")

    longestGeneCoordinates = list()
    currentLine = fileIN.readline().rstrip().split('\t') 
    currentGeneCoordinates = [currentLine[0], float(currentLine[1]), currentLine[2], currentLine[3], currentLine[4], currentLine[5], currentLine[6], currentLine[7], currentLine[8], currentLine[9]]   

    for counter, i in enumerate(fileIN): #Go through each line
        i = i.rstrip().split("\t")
        i = [i[0], float(i[1]), i[2], i[3], i[4], i[5], i[6], i[7], i[8], i[9]]
        if i[0] == currentGeneCoordinates[0]:
            currentPos = float(currentGeneCoordinates[1])
            thisPos = float(i[1])
            if thisPos < currentPos:
                currentGeneCoordinates = i
                
                
        if i[0] != currentGeneCoordinates[0]: 
            #If this line's gene does not match the currentGene's gene, then we must have moved on to the next gene.
            #Since we're now at a new gene, let's store the currentGene
            longestGeneCoordinates.append(currentGeneCoordinates)
            #Also, let's update the currentGene to match this gene. 
            currentGeneCoordinates = i
        if counter %100000==0:
            print(counter)

    
    fileOUT = open("ldsc_input.txt", "w")
    for i in longestGeneCoordinates:
        print(i)
        fileOUT.write("\t".join([str(x) for x in i]) + "\n")  #Join will take a list, and create ONE string. where each element of the list is separated by the character in the quotation marks (here it is a "\t")
        #Also,we're converting the items into a string - because we can't output integers. 
    fileOUT.close() 
    fileIN.close
output_ldsc()
```
